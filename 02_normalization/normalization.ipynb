```python
data_read = pd.read_excel('./Concrete_Data.xls')

features = data_read.columns
data = data_read.values

print('Dimensions of the Dataset: ' + str(np.shape(data)))
```

```python
X, Y = data[:, 0:8], data[:, 8]
# X: Input parameters or features, Y: recorded output value
```

```python
data_mean = np.mean(X, axis = 0)
data_var = np.var(X, axis = 0)

print('Dimensions of data_mean: ' + str(np.shape(data_mean)))
print('Dimensions of data_var: ' + str(np.shape(data_var)))
```

```python
# Z score normalization

X_norm = (X -  data_mean) / np.sqrt(data_var)
print('First row of X_norm: ' + str(np.ndarray.tolist(X_norm[0,])))

Y_norm = (Y - np.mean(Y)) / np.sqrt(np.var(Y))
print('First element of Y_norm: ' + str(Y_norm[0]))
```

```python
Y.shape
```

```python
combined = np.c_[X_norm, Y]

# Reference element at row 1, column 2
print('combined[1, 2]) = ' + str(combined[1, 2]))

print('Shape of normalized data: ' + str(np.shape(combined)))

combined_frame = pd.DataFrame(combined, columns = features)
```

```python
# With a Python list

combined_list = np.ndarray.tolist(combined)

# Reference element at row 1, column 2
print('combined([1][2]) = ' + str(combined[1][2]))

combined_frame = pd.DataFrame(combined, columns = features)
```

```python
# Displaying dataframe
combined_frame
```

```python
# Saving data as a csv file
combined_frame.to_csv('z_score_norm.csv', index = False)
```

```python
np.shape(X)[0]
```

```python
scaled_X = None

show_shape = True

for index in range(np.shape(X)[1]):
    # index holds the column index
    column = X[:, index]
    
    scaler = MinMaxScaler()
    
    points = column.reshape(-1, 1)
    scaler.fit(points)
    scaled = scaler.transform(points)
    
    if scaled_X is None:
        scaled_X = scaled
    else:
        scaled_X = np.c_[scaled_X, scaled]
    
    if show_shape:
        print('Shape now: ' + str(scaled_X.shape))
```

```python
combined_scaled = np.c_[scaled_X, Y]

combined_scaled_frame = pd.DataFrame(combined_scaled, columns = features)

# Displaying dataframe
combined_scaled_frame
```

```python
# Saving data as a csv file
combined_scaled_frame.to_csv('minMax_norm.csv', index = False)
```

```python
# Decimal scaling

def get_oom(points):
    # Returns order of magnitude of the largest value
    
    highest = max(abs(points))
    highest = int(highest)
    
    magnitude = len(str(highest))
    return magnitude

deci_X = None

show_mag = True

for index in range(np.shape(X)[1]):
    # index holds the column index
    column = X[:, index]
    
    magnitude = get_oom(column)
    
    if show_mag:
        print('Column: ' + str(index + 1) + ', OOM: ' + str(magnitude))
    
    denominator = np.power(10, magnitude)
    
    deci_col = column / denominator
    
    if deci_X is None:
        deci_X = deci_col
    else:
        deci_X = np.c_[deci_X, deci_col]
        
combined_deci = np.c_[deci_X, Y]

combined_deci_frame = pd.DataFrame(combined_deci, columns = features)
```

```python
# Displaying dataframe
combined_deci_frame
```

```python
# Saving decimal scaled data as a csv file
combined_deci_frame.to_csv('deci_norm.csv', index = False)
```
